#### Model Selection
Since many regression model are provided, we do cross-validation to select the best among others.

The important parameters to consider is:
- scoring: Since the scoring is on "Regression" model, scikit-learn provides many scoring for us. 
- cv: This parameter tells how many folds in cross validation. It's a integer value and it should be greater than 1.

Combine all GridSearch (for best hyper of each model) and Cross-validation (for try that best hyper-tuned model) -> GridSearchCV
1. define all algorithms, params wanted to tune, and cv strategy (kfold) must not transformed before pass to CV!
2. GridSearchCV to tune hyperparameters of that model -> best model
3. Cross-validation that best model

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor, HistGradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor

from sklearn.model_selection import KFold, cross_val_score

algorithms = {
    'Linear Regression': LinearRegression(),
    'Ridge': Ridge(random_state=0),
    'Lasso': Lasso(random_state=0),
    'Gradient Boosting': GradientBoostingRegressor(random_state=0),
    'Histogram-based Gradient Boosting Regression Tree': HistGradientBoostingRegressor(random_state=0),
    'Epsilon-supported Vector Regression': SVR(),
    'KNeighbors Regression': KNeighborsRegressor(),
    'Decision Tree Regression': DecisionTreeRegressor(random_state = 0),
    'Random Forest Regression': RandomForestRegressor(random_state = 0),
    'AdaBoost': AdaBoostRegressor(random_state=0)
    }

param_grids = {
    'Linear Regression': {},
    'Ridge': {},
    'Lasso': {},
    'Gradient Boosting': {},
    'Histogram-based Gradient Boosting Regression Tree': {},
    'Epsilon-supported Vector Regression': {},
    'KNeighbors Regression': {},
    'Decision Tree Regression': {},
    'Random Forest Regression': {},
    'AdaBoost': {}
    }

# defining splits
kfold = KFold(n_splits=5, shuffle=True)

# Scoring for regression
score_rule = 'r2'
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import GridSearchCV

# Create a dictionary to store the best model for each fold
all_models = {}

# Perform model selection using cross-validation
for model_name, model in algorithms.items():
    
    print(f"GridSearchCV on {model_name}")
    print("Generating pipeline...")
    grid_pipe = Pipeline(steps=[('preprocessor', preprocessor),
                                ('estimator', model)])
    

    # print('old_dict (old_param): ', param_grids[model_name])
    new_param = dict((f"{grid_pipe.steps[-1][0]}__{k}",v) for k,v in param_grids[model_name].items())

    print('Tuned following hyperparams: ', new_param, end='  \t')

    # Use GridSearchCV to perform hyperparameter tuning and select the best model
    # No information leakage
    grid_search = GridSearchCV(estimator = grid_pipe,
                               param_grid = new_param,
                               cv=kfold,
                               scoring=score_rule,
                               n_jobs=-1,
                               return_train_score=True,
                               refit=True)
    
    print("Tuned done.")

    print('Fitting data...',end='\t')
    # Fit the model on the training data
    grid_search.fit(X_train, y_train)
    print('Done!')
    print('Mean train score is:', grid_search.cv_results_['mean_train_score'])
    print('Mean test score is:', grid_search.cv_results_['mean_test_score'])
    print(f"Time taken: {(grid_search.cv_results_['mean_fit_time']+grid_search.cv_results_['mean_score_time'])*5} s")
    # store all models results
    all_models[model_name] = {'best_estimator_': grid_search.best_estimator_,
                              'cv_results_': grid_search.cv_results_}
```

```python
list_keep = []
for name, sth in all_models.items():
    print(f"{name}")
    print(f"Time taken: {sth['cv_results_']['mean_fit_time']*5} seconds")
    print(f"Mean train score: {sth['cv_results_']['mean_train_score']}")
    print(f"Mean validate score: {sth['cv_results_']['mean_test_score']}")
    list_keep.append([name, sth['cv_results_']['mean_test_score'].min()])
```

```python
# added AQI columns based on US AQI levels
def get_AQI(pm_value):

    if pm_value < 0:
        AQI = np.nan
    elif pm_value <= 12:
        AQI = 'Good'
    elif pm_value <= 35.4:
        AQI = 'Moderate'
    elif pm_value <= 55.4:
        AQI = 'Unhealthy for Sensitive Group'
    elif pm_value <= 150.4:
        AQI = 'Unhealthy'
    elif pm_value <= 250.4:
        AQI = 'Very Unhealthy'
    else:
        AQI = 'Hazardous'

    return AQI
```
